```python
import requests
from bs4 import BeautifulSoup
import time
import csv

# Specify a delay (in seconds)
delay = 0.5

# Make a request to the website
rq_1 = requests.get('https://incidecoder.com/ingredient-functions')

# Parse the HTML content of the page with BeautifulSoup
sp = BeautifulSoup(rq_1.text, 'html.parser')

# Get all ingredient divs
ingredient_divs = sp.find('div', class_='paddingtbl').find_all('a')

ingredient_class = []
# Print the text for each link found within the div
for ingredient in ingredient_divs:
    ingredient_with_dash = ingredient.text.strip().replace(' ', '-').replace('/','-')
    ingredient_class.append(ingredient_with_dash)


ingredients = []

for i in ingredient_class:
    new_url = f'https://incidecoder.com/ingredient-functions/{i}'
    classification = i
    print(i)
    print(new_url)

    offset = 0
    while True:
        url_to_request = f'{new_url}?offset={offset}'
        # Get the page
        rq_2 = requests.get(url_to_request)
        soup = BeautifulSoup(rq_2.text, 'html.parser')

        # Get the ingredients on this page
        ingredient_divs_1 = soup.find('div', class_='paddingtbl')
        ingredient_links = ingredient_divs_1.find_all('a') if ingredient_divs_1 else []
        if not ingredient_links:  # If there are no ingredient links, stop the loop
            break
        for ingredient_1 in ingredient_links:
            name = ingredient_1.text.strip().replace(' ', '-')
            ingredients.append((classification, name))

        # Increment the offset for the next page
        offset += 1

        # Delay before the next request
        time.sleep(1)

    # Print the list of ingredients
    #print(ingredients[:5])


with open('ingredient_classification.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Classification", "Name"])  # write header
    writer.writerows(ingredients)
```
